```python
import spacy
nlp = spacy.load('en_core_web_md')
movie_genre = ['Anime','Romance','Comedy','History','Action','Horror','Drama','Mystery','Thriller','Science Fiction','Fantasy']
```

```python
musicgen = "happy birthday to you & me"
musicgen.split(' ')[0]
```

```python
def books_to_movie(musicgen1):
  similar_genres = []
  for movie in movie_genre:

    words = musicgen1 + ' ' + movie
    tokens = nlp(words)

    # token1, token2, token3 = tokens[0], tokens[1], tokens[2]  

    token1, token2 = tokens[0], tokens[1]  
    similarityval1 = token1.similarity(token2) 
    # similar_genres.append([movie, similarityval])
    # similarityval2 = token2.similarity(token3) 
    # if similarityval1 > similarityval2:
    similar_genres.append([movie,similarityval1])
    # else:
    #   similar_genres.append([movie,similarityval2])
  similar_genres = sorted(similar_genres, key = lambda x: x[1], reverse=True)
  similar_genres = similar_genres[:4]
  result = [ x[0] for x in similar_genres]
  return result
```

```python
res = books_to_movie('Humour')
print(res)
```

```python
def map_genre_helper(musicgen1):
  res = []
  if '&' in musicgen1:
      word1 = musicgen1.split(' ')[0]
      word2 = musicgen1.split(' ')[2]
      res1 = books_to_movie(word1)
      res2 = books_to_movie(word2)
      res = res1[:2]
      for r in res2:
        if (r not in res) & (len(res)<4):
          res.append(r)
  else:
    res = books_to_movie(musicgen1)
  return res
```

```python
def map_genre():
  gen_list = items.columns
  gen_list = gen_list[2:]
  for g in gen_list:
    gen_dict[g] = map_genre_helper(g)
```
