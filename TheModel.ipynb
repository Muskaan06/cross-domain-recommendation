# recomm code

```python
user_item_rating.columns = ['item_id', 'user_id', 'rating']
```

```python
user_item_matrix = user_item_rating.pivot_table(index = 'item_id', columns ='user_id', values = 'rating')
user_item_matrix
```

```python
def item_similarity(item1,item2):
    sim = item_df.at[item1,item2]
    return sim
```

```python
users = user_item_matrix.columns.to_list()
len(users)
```

```python
artist_list = user_item_matrix.index.tolist()
len(artist_list)
```

```python
#custom function to filter the seen movies and unseen movies of the target user

def target_movies_to_users(target_person):
    target_person_movie_lst = []
    unique_list = item_list
    # for movies in dataset[target_person]:
    #     target_person_movie_lst.append(movies)
    target_person_movie_lst = user_item_matrix[target_person].dropna().index.values.tolist()

    s=set(unique_list)
    recommended_movies=list(s.difference(target_person_movie_lst))
    a = len(recommended_movies)
    if a == 0:
        return 0
    return recommended_movies,target_person_movie_lst
```

```python

def recommendation_phase(target_person):

    if target_movies_to_users(target_person=target_person) == 0:
        print(target_person, "has seen all the movies")
        return -1
    not_seen_movies,seen_movies=target_movies_to_users(target_person=target_person)
# 
    # seen_ratings = [[dataset[target_person][movies],movies] for movies in dataset[target_person]]

    rating_value = user_item_matrix[target_person].dropna()
    seen_ratings = [[rating_value[movies],movies] for movies in rating_value.index.values.tolist()]
    print('------------')
# 
    # seen_ratings = sorted(seen_ratings,key=lambda x: (x[0]), reverse=True)

    print(seen_ratings)
# 
    for rate,movie in seen_ratings[:5]:
      not_seen_movies.append(movie)
      test_list.append(movie)
      actual.append(rate)
      print("TEST: ",movie)
    seen_ratings = seen_ratings[5:]

    rankings =[]
# 
    # for artist in artist_list:
    #   not_seen_movies.append(artist)

    for i in not_seen_movies:
        weighted_avg,weighted_sim = 0,0
        sum_r, num_r = 0,0
        for rate,movie in seen_ratings:
            sum_r+=rate
            num_r+=1
            # print("first-- ", i, " second-- ", movie)
            # if(i=='M0005000009' or movie=='M0005000009'):
            #   print("first-- ", i, " second-- ", movie)
            item_sim=0
            try:
              item_sim=item_similarity(i,movie)
            except:
              item_sim=0
            if item_sim>np.float64(0.95):
              weighted_avg +=(item_sim*rate)
              weighted_sim +=item_sim
        if weighted_sim==0:
          weighted_rank = sum_r/num_r   #  ASK WHAT IS CORRECT HERE
        else:
          weighted_rank=weighted_avg/weighted_sim
        rankings.append([weighted_rank,i])

    # rankings.sort(reverse=True)
    # return rankings[:1000]
    return rankings
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import confusion_matrix
from math import sqrt
```

```python
# print("Enter the target person/user id")
# # tp = input().title()
# tp = int(input())

testvalue_list = []
num_commons = []
rmse_scores = []
mae_scores = []
for user in users:
    value_list = []
    value_list.append(user)
    rec_list=[]
    test_list=[]
    actual=[]
    pred=[]
    a=recommendation_phase(user)
    if a != -1:
        print("Recommendation Using Item based Collaborative Filtering:  ")
        for rate,movie in a:
          # if(rate>=4):  #rate variable here means predicted rating
          #   rec_list.append(movie)
            #print(rate," ---> ",movie)
            if(movie in test_list):
              rec_list.append([rate,movie])
              pred.append(rate)
        print(rec_list)
        pred = pred[:5]

        #calculate RMSE
        rmse = sqrt(mean_squared_error(actual, pred))
        value_list.append(rmse)
        rmse_scores.append(rmse)
        print('RMSE: ',rmse)

        #calculate MAE
        mae = mean_absolute_error(actual,pred)
        value_list.append(mae)
        mae_scores.append(mae)
        print('MAE: ',mae)

        #calculate hit ratio 
        hit = 0
        for i in range(len(actual)):
          diff = actual[i] - pred[i]
          if diff < 0.5 and diff > -0.5:
            hit = hit + 1
        hit_ratio = hit/len(actual)
        print('Hit Ratio: ', hit_ratio)
        value_list.append(hit_ratio)


        #calculate confusion matrix to find precision, recall etc.
        y_true = []
        y_pred = []

        for item in actual:
          if item >= 3:
            y_true.append(1)
          else:
            y_true.append(0)

        for item in pred:
          if item >= 3:
            y_pred.append(1)
          else:
            y_pred.append(0)
            
        tn, fp, fn, tp = confusion_matrix(y_true, y_pred, labels=[0, 1]).ravel()
        # cm = confusion_matrix(y_true, y_pred, labels=[0, 1])
        # print("confusion : ", cm)
        
        recall = tp / (tp + fn)       # definition of recall
        precision = tp / (tp + fp)    # definition of precision
        f1_score = 2*recall*precision/(recall+precision)

        print("precision :", precision)
        print("recall :", recall)
        print("f1 score :", f1_score)
        
        value_list.append(precision)
        value_list.append(recall)
        value_list.append(f1_score)

        # print(set(rec_list) & set(test_list))
        # common = set(rec_list) & set(test_list)
        # num_commons.append(len(common))
        testvalue_list.append(value_list)

else:
    print("Person not found in the dataset..please try again")

test_values = pd.DataFrame(testvalue_list, columns = ["user_id", "rmse", "mae", "hit ratio", "precision", "recall", "f1 score"])
test_values
```

```python

def recommendation_phase(target_person):

    if target_movies_to_users(target_person=target_person) == 0:
        print(target_person, "has seen all the movies")
        return -1
    not_seen_movies,seen_movies=target_movies_to_users(target_person=target_person)
# 
    # seen_ratings = [[dataset[target_person][movies],movies] for movies in dataset[target_person]]

    rating_value = user_item_matrix[target_person].dropna()
    seen_ratings = [[rating_value[movies],movies] for movies in rating_value.index.values.tolist()]
    print('------------')
# 
    # seen_ratings = sorted(seen_ratings,key=lambda x: (x[0]), reverse=True)

    print(seen_ratings)
# 
    movie_count = 0
    kindle_count = 0

    # for rate,movie in seen_ratings:
      
    #   if movie_count < 5 and movie in item_movie:
    #     movie_count = movie_count+1
    #     not_seen_movies.append(movie)
    #     test_list.append(movie)
    #     actual.append(rate)
    #     print("TEST: ",movie)
    #     seen_ratings.remove([rate,movie])
      # elif kindle_count < 3 and movie in item_kindle:
      #   kindle_count = kindle_count+1
      #   not_seen_movies.append(movie)
      #   test_list.append(movie)
      #   actual.append(rate)
      #   print("TEST: ",movie)
      #   seen_ratings.remove([rate,movie])
      
      
    # seen_ratings = seen_ratings[5:]

    test_list_full = []
    actual_full = []



    for rate,movie in seen_ratings:
      if movie in item_movie:
        test_list_full.append(movie)
        actual_full.append(rate)
        # print('TEST: ', movie)
        seen_ratings.remove([rate,movie])


    for i in range(5):
      test_list.append(test_list_full[i])
      actual.append(actual_full[i])



    rankings =[]
# 
    for artist in artist_list:
      not_seen_movies.append(artist)

    for i in not_seen_movies:
        weighted_avg,weighted_sim = 0,0
        sum_r, num_r = 0,0
        for rate,movie in seen_ratings:
            sum_r+=rate
            num_r+=1
            # print("first-- ", i, " second-- ", movie)
            # if(i=='M0005000009' or movie=='M0005000009'):
            #   print("first-- ", i, " second-- ", movie)
            item_sim=0
            try:
              item_sim=item_similarity(i,movie)
            except:
              item_sim=0
            if item_sim>np.float64(0.5):
              weighted_avg +=(item_sim*rate)
              weighted_sim +=item_sim
        if weighted_sim==0:
          weighted_rank = sum_r/num_r   #  ASK WHAT IS CORRECT HERE
        else:
          weighted_rank=weighted_avg/weighted_sim
        rankings.append([weighted_rank,i])

    # rankings.sort(reverse=True)
    # return rankings[:1000]
    return rankings
```
